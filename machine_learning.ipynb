```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
import sklearn as sk
```

```python
df = pd.read_csv('heart_disease.csv')
df.head() 
```

```python
# Show number of people with heart diseases
# Column target
df.target.value_counts()
```

```python
# Show percetge of male and female
count_female = len(df[df.sex == 0])
count_male = len(df[df.sex == 1])
#Show Output
print('Percentage of female patients: {:.2f}%'.format(count_female/(len(df.sex))*100))
print('Percentage of male patients: {:.2f}%'.format(count_male/(len(df.sex))*100))
```

```python
# Mean
df.groupby('target').mean()
```

```python
# Bar plot showing heart disease frequency od age
pd.crosstab(df.age, df.target).plot(kind='bar',figsize=(20,6))
pl.title('Heart Disease Frequency for Ages')
pl.xlabel('Age')
pl.ylabel('Frequency')
pl.show()
```

```python
# Data Visualization for Heart Disease Frequency For Sex
pd.crosstab(df.sex, df.target).plot(kind='bar',figsize=(20,6), color =['#6F8FAF','#228B22'])
pl.title('Heart Disease Frequency for Sex')
pl.xlabel('Sex (0 = Female, 1 = Male)')
pl.legend('Has Disease','DoesNot have Disease')
pl.ylabel('Frequency')
pl.xticks(rotation=0)
pl.show()
```

```python
# Scatter plot showing Cholestrol and age
pl.scatter(df.age, df.chol)
pl.title('Scatter plot Showing Blood Cholestrol for various Ages')
pl.xlabel('Age')
pl.ylabel('Cholestrol')
pl.show()
```

```python
# Workingwith ML model accuracy of different parameters
# Creating a Model for Logistic regression
y = df.target.values
x_data = df.drop(['target'],axis=1)
```

```python
#Normalizing Data( Splitting Data to training and testing)
# 80% training and 20% testing
# Normalize the data X = (X-Xmin)/(Xmax-Xmin)
X = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
```

```python
from sklearn.model_selection import train_test_split

# Assuming you have features X and labels y
```

```python
# Evaluating Performance Accuracy 100% ! 99.87%
# Model testing( Be specificwith the training algorithms ie linear regression, decision trees,svm, knn etc)
# Model Evaluation( accracy, precision, recall,F1score,mean squared error)
# Performance Insights(Insight about the performance)
```

```python
# Import ML Library
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Dataset.Let X be the independent variables( input features) and Y to be the dependent variables
# Your target variable(target variable)
# With logistic regression, it is important to store independent variables in x.
# Store dependent variables i(target data) in Y.

# Split data into training and testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)

#Transpose Matrices
x_train = X_train.T
y_train = y_train.T
x_test = X_test.T
y_test = y_test.T

# Model Training
model = LogisticRegression()
model.fit(X_train, y_train)

# Model Evaluation
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# Display results
print('Accuracy: ', accuracy)                       
```
